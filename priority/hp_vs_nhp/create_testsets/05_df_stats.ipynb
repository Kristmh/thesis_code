```python
import pandas as pd
# import csv file
df = pd.read_csv('csv/flutter_testset.csv')
df
```

```python
# Make a new df with only the labels that contains: P0, P1, P2, P3
df_p0 = df[df['labels'].str.contains('P0')]
df_p1 = df[df['labels'].str.contains('P1')]
df_p2 = df[df['labels'].str.contains('P2')]
df_p3 = df[df['labels'].str.contains('P3')]
df_p0["class"] = "P0"
df_p1["class"] = "P1"
df_p2["class"] = "P2"
df_p3["class"] = "P3"
# Concatenate the dataframes
df_priority = pd.concat([df_p0, df_p1, df_p2, df_p3])
df_priority
```

```python
# Count labels
df_priority['class'].value_counts().to_frame()[:50]
```
