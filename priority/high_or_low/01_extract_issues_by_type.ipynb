```python
import pandas as pd
import re
import os
```

```python

# folder path
dir_path = r'../../csv/all_issues' ## Point to the extracted folder containing all the issues csv files

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

```python
appended_data = []
# Regular expression to capture various variations of "high priority"
# high_priority = r"\b(?:priority[\s\-_,.:]*high|high[\s\-_,.:]*priority|top[\s\-_,.:]*priority|high[\s\-_,.:]*prio|high[\s\-_,.:]*pri|prio[\s\-_,.:]*high|pri[\s\-_,.:]*high)\b"
high_priority = r"\bhigh\W*p(?:ri(?:o(?:rity)?)?)?\b|\bp(?:ri(?:o(?:rity)?)?)?\W*high\b"

medium_priority = r"\b(?:priority[\s\-_,.]*medium|medium[\s\-_,.]*priority|medium[\s\-_,.]*prio|medium[\s\-_,.]*pri|prio[\s\-_,.]*medium|pri[\s\-_,.]*medium)\b"

#low_priority = r"\b(?:priority[\s\-_,.]*low|low[\s\-_,.]*priority|low[\s\-_,.]*prio|low[\s\-_,.]*pri|prio[\s\-_,.]*low|pri[\s\-_,.]*low)\b"
low_priority = r"\blow\W*p(?:ri(?:o(?:rity)?)?)?\b|\bp(?:ri(?:o(?:rity)?)?)?\W*low\b"

pattern=high_priority

for i in res:
    try:
        file_path = f"{dir_path}/{i}"
        df = pd.read_csv(file_path, index_col=0)

        filtered_df = df[
           df["labels"].str.contains(pattern, case=False, na=False, regex=True)
        ]

        if not filtered_df.empty:  # Append non-empty dataframes to the list
           appended_data.append(filtered_df)


        print(f"Processed: {i}")
    except Exception as e:
        print(f"Error processing {i}: {e}")


appended_data = pd.concat(appended_data, ignore_index=True)
```
