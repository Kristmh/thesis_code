```python
def filter_texts_by_length(df, column_name='body', min_length=1000):
    """
    Filters a DataFrame to keep only rows where the specified column's text length is at least the given minimum.

    Parameters:
    - df: pandas.DataFrame, the DataFrame to filter.
    - column_name: str, the name of the column containing text to check length.
    - min_length: int, the minimum length of text to keep.

    Returns:
    - pandas.DataFrame, filtered DataFrame.
    """
    length_before = len(df)
    # Calculate the length of each string in the specified column
    df['text_length'] = df[column_name].str.len()

    # Filter the DataFrame to keep only rows where text length meets the minimum requirement
    filtered_df = df[df['text_length'] >= min_length]

    # Drop the temporary 'text_length' column
    filtered_df = filtered_df.drop(columns=['text_length'])
    length_after = len(filtered_df)
    print(f"Number of rows before: {length_before}")
    print(f"Number of rows after: {length_after}")
    print(f"Difference: {length_before - length_after}")

    return filtered_df
```

```python
# Make a new dataframe with only text, label and class cols.
all_priority_subset = all_priority[["text" , "label" , "class"]]
all_priority_subset
```

```python
# Clean the data.
all_priority_subset["text_clean"] = all_priority_subset["text_str"].map(preprocess_text)
```

```python
# Make a subset with text_clean and label
priority_label_text = all_priority_subset[["text_clean" , "label"]]
priority_label_text
```

```python
# Need to dropna here since cleaning function returns NaN for not english text.
priority_label_text.dropna(inplace=True)
priority_label_text.reset_index(inplace=True)
priority_label_text.drop(columns=["index"] , inplace= True)

priority_label_text
```

```python
priority_label_text["label"].value_counts()
```

```python
from sklearn.utils import resample

# Split the dataset into two based on the label
df_majority = priority_label_text[priority_label_text['label'] == 0]
df_minority = priority_label_text[priority_label_text['label'] == 1]

# Undersample the majority class
df_majority_undersampled = resample(df_majority, 
                                     replace=False,    # sample without replacement
                                     n_samples=len(df_minority),     # to match minority class
                                     random_state=123) # reproducible results

# Combine minority class with downsampled majority class
df_balanced = pd.concat([df_majority_undersampled, df_minority])

# Shuffle the dataset to avoid any ordering bias
df_balanced = df_balanced.sample(frac=1).reset_index(drop=True)
priority_label_text = df_balanced
priority_label_text["label"].value_counts()
```

```python
# Splitting the DataFrame
test_df = priority_label_text.sample(frac=0.05, random_state=1)  # Select 5% of the data
test_file_name = f"csv/clean_test_high_vs_random.csv"
test_df.to_csv(test_file_name, index=False)
priority_df = priority_label_text.drop(test_df.index)
priority_df.reset_index(drop=True, inplace=True)
priority_df
```

```python
# Clean dataset with clean text and labels.
# 0 = high priority, 1 = not high priority
file_name = f"csv/clean_high_vs_random.csv"
priority_df.to_csv(file_name)
```
