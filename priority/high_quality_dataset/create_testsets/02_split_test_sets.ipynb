```python
import pandas as pd
```

```python
df = pd.read_csv('csv/test_repos.csv')
# Remove column named Unnamed:0
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
# Define the pattern as a variable
pattern = 'priority-0-high'

# Check if 'labels' contains the pattern
filtered_df = df[df['labels'].str.contains(pattern)]
# Reset index
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
# Split dataframe into the different test repos
mypy = df[df['repo'] == 'python/mypy']
mypy = mypy.reset_index(drop=True)
rust = df[df['repo'] == 'rust-lang/rust']
rust = rust.reset_index(drop=True)
flutter = df[df['repo'] == 'flutter/flutter']
# Reset index  
flutter = flutter.reset_index(drop=True)
flutter
```

```python
# Save the filtered dataframes to csv
mypy.to_csv('csv/mypy_testset.csv',index=False)
rust.to_csv('csv/rust_testset.csv', index=False)
flutter.to_csv('csv/flutter_testset.csv', index=False)
# Read the csv files to check if they are saved correctly
mypy = pd.read_csv('csv/mypy_testset.csv')
rust = pd.read_csv('csv/rust_testset.csv')
flutter = pd.read_csv('csv/flutter_testset.csv')
```

```python
# Checking how many high priority issues are in the rust repo
pattern = 'P-(high|critical)'

# Check if 'labels' contains the pattern
rust = rust[rust['labels'].str.contains(pattern)]
# Reset index
rust = rust.reset_index(drop=True)
rust
```

```python
pattern = 'P(1|0)'
# Check if 'labels' contains the pattern
hp_flutter = flutter[flutter['labels'].str.contains(pattern)]
# Reset index
hp_flutter = hp_flutter.reset_index(drop=True)
hp_flutter
```

```python
pattern = "priority-0-high"
# Check if 'labels' contains the pattern
hp_mypy = mypy[mypy['labels'].str.contains(pattern)]
# Reset index
hp_mypy = hp_mypy.reset_index(drop=True)
hp_mypy
```
