# Dataset creation
This notebook will:
* Remove Technical debt labels
* Extract 5% to make testing set.
* It will produce 3 different dataset:
  1. High/med/priority dataset with no debt
  2. Test dataset
  3. Debt dataset

```python
import pandas as pd
```

```python
# Read in the full dataset
file_name = "random_per_repo"
full_file_name = f'csv/{file_name}.csv'
random = pd.read_csv(full_file_name, index_col=0)
random
```

```python
# Create debt dataframe
technical_debt_regex = r'debt|\bTD\b'
contains_debt_high_priority = random['labels'].str.contains(technical_debt_regex, case=False, na=False)
debt = random[contains_debt_high_priority].reset_index(drop=True)
debt
```

```python
debt.labels.value_counts().to_frame()[:50]
```

```python
# Dataset with debt and the given priority
debt_file_name = f"csv/{file_name}_with_td.csv"
debt.to_csv(debt_file_name)
```

```python
# To labels
random.labels.value_counts().to_frame()[:50] 
```

```python
# Remove debt from the dataset
priority_no_debt = random[~contains_debt_high_priority].reset_index(drop=True)
priority_no_debt.reset_index(drop=True, inplace=True)
priority_no_debt
```

```python
# Sanity check
should_be_empty=priority_no_debt[priority_no_debt['labels'].str.contains(technical_debt_regex, case=False, na=False)]
should_be_empty
```

```python
# Convert to csv
priority_file_name = f"csv/{file_name}_no_td.csv"
priority_no_debt.to_csv(priority_file_name)
```
