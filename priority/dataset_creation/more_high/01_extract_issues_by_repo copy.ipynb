# Dataset creation
This notebook will:
* Find github issues by label

```python
import pandas as pd
import re
import os
```

```python
df = pd.read_csv('csv/high_priority_full_dataset.csv')
# Get the unique repo names as a python list
repo_names = df['repo'].unique().tolist()
repo_names
```

```python
# folder path
dir_path = r'../../../csv/all_issues' ## Point to the extracted folder containing all the issues csv files

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

```python
length_res = len(res)
appended_data = []
for i, r in enumerate(res):
    if i > 2:
        break
    try:
        file_path = f"{dir_path}/{r}"
        df = pd.read_csv(file_path, index_col=0)

    
        # Make sure the dataframe is not empty
        df = df[df['labels'].notnull() & df['labels'].str.strip().astype(bool)]

        for repos in repo_names:
            temp_df = df[df['repo'] == repos]
            if not df.empty:  # Append non-empty dataframes to the list
                appended_data.append(temp_df)


        print(f"{i}/{length_res} Processed: {r} ")
    except Exception as e:
        print(f"{i}/{length_res} Error processing {r}: {e}")

appended_data = pd.concat(appended_data, ignore_index=True)
```

```python
appended_data = pd.DataFrame(appended_data)
appended_data
```
