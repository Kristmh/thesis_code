# Data cleaning
We want the class with labels as numerical value and the body with clean text.

This will remove:
* duplicates
* NaN entires
* non english
* url, html

* make it lowercase
* combine title and body

```python
import pandas as pd
import sys
sys.path.append("../../../scripts_shared/")
from preprocess_text import preprocess_text
```

```python
# Read CSV into a dataframe
high_priority = pd.read_csv("csv/high_priority_full_dataset.csv", index_col=0)

high_priority
```

```python
# Remove debt from datasets
technical_debt_regex = r'debt|\bTD\b'
contains_debt_high_priority = high_priority['labels'].str.contains(technical_debt_regex, case=False, na=False)
debt = high_priority[contains_debt_high_priority].reset_index(drop=True)
debt
```

```python
# Remove debt from the dataset
high_priority = high_priority[~contains_debt_high_priority].reset_index(drop=True)
high_priority.reset_index(drop=True, inplace=True)
high_priority
```

```python
# Number or different labels
high_priority.labels.value_counts().to_frame()[:50]
```

```python
high_priority[high_priority["repo"] == "python/mypy"]
```

```python
# Remove mypy from the dataset
high_priority = high_priority[high_priority["repo"] != "python/mypy"]
high_priority = high_priority[high_priority["repo"] != "rust-lang/rust"]
high_priority = high_priority[high_priority["repo"] != "flutter/flutter"]
# reset index
high_priority.reset_index(drop=True, inplace=True)
high_priority
```

```python
# Double check that test repos are removed.
high_priority[high_priority["repo"] == "python/mypy"]
```

```python
high_priority[high_priority["repo"] == "rust-lang/rust"]
```

```python
high_priority[high_priority["repo"] == "flutter/flutter"]
```

```python
#Give each priority a label by number.
# 'Label encoding'. Makes is easier for machine learning models to work with categorical data.
high_priority["label"] = 1
high_priority["class"] = "high_priority"
high_priority.head()
```

```python
# Drop duplicates by the content of the title
high_priority = high_priority.drop_duplicates(subset=['title'], keep='last')
high_priority.dropna(inplace=True)
high_priority.reset_index(inplace=True)
high_priority.drop(columns=["index"] , inplace= True)
high_priority["class"].value_counts()
```

```python
high_priority.info()
```

```python
# Value counts before cleaning.
high_priority["labels"].value_counts().to_frame()[:50]
```

```python
print(high_priority["title"][0])
print(high_priority["body"][0])
```

```python
# Copy content of body to a new col named text
high_priority["text"] = high_priority["title"] + " " + high_priority["body"]
high_priority.tail()
```

```python
high_priority["text"][0]
```

```python
# Make a new dataframe with only text, label and class cols.
all_priority_subset = high_priority[["text" , "label" , "class", "repo", "labels"]]
all_priority_subset
```

```python
# Convert to string
all_priority_subset["text"] = all_priority_subset['text'].astype(str)
```

```python
all_priority_subset
```

```python
# Clean the data.
all_priority_subset["text_clean"] = all_priority_subset["text"].map(preprocess_text)
```

```python
# Make a subset with text_clean and label
priority_label_text = all_priority_subset[["text_clean" , "label", "class", "repo", "labels"]]
priority_label_text
```

```python
# Row with NaN
priority_label_text[priority_label_text.isna().any(axis=1)]
```

```python
# Need to dropna here since cleaning function returns NaN for not english text.
priority_label_text.dropna(inplace=True)
priority_label_text.reset_index(drop=True,inplace=True)

priority_label_text
```

```python
# Rows with NaN
priority_label_text[priority_label_text.isna().any(axis=1)]
```

```python
# Drop rows with null
priority_label_text.dropna(inplace=True)
priority_label_text.reset_index(drop=True,inplace=True)
priority_label_text
```

```python
priority_label_text[priority_label_text.isna().any(axis=1)]
```

```python
null_rows = priority_label_text[priority_label_text['text_clean'].isnull()]
null_rows
```
