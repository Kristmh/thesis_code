```python
import pandas as pd
```

```python
# Read CSV into a dataframe
high_priority = pd.read_csv("csv/final_high_priority_dataset.csv", index_col=0)
not_high_priority = pd.read_csv("csv/final_not_high_priority_dataset.csv", index_col=0)
```

```python
high_priority
```

```python
# Number or different labels
high_priority.labels.value_counts().to_frame()[:50]
```

```python
# Number or different labels
not_high_priority.labels.value_counts().to_frame()[:50]
```

```python
# Add a new column
high_priority["class"] = "high_priority"
high_priority
```

```python
not_high_priority["class"] = "not_high_priority"
not_high_priority
```

```python
all_data = pd.concat([high_priority, not_high_priority] , ignore_index= True)
```

```python
all_data
```

```python
all_data_no_dupes = all_data.drop_duplicates()
all_data_no_dupes
```

```python
# Convert dataframe to a CSV file.
all_data_no_dupes.to_csv("csv/full_dataset_priority_high_or_not_high.csv")
```
