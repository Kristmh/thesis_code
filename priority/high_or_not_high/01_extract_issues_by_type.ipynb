```python
import pandas as pd
import re
import os
```

```python

# folder path
dir_path = r'../../csv/all_issue_event_1' ## Point to the extracted folder containing all the issues csv files

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

```python
appended_data = []
not_high_priority = []
# Regular expression to capture various variations of "high priority"
pattern = r"\b(?:priority[\s\-_,.]*high|high[\s\-_,.]*priority|top[\s\-_,.]*priority|urgent|critical|high[\s\-_,.]*prio|high[\s\-_,.]*pri|prio[\s\-_,.]*high|pri[\s\-_,.]*high)\b"


for i in res:
    try:
        file_path = f"{dir_path}/{i}"
        df = pd.read_csv(file_path, index_col=0)

        # filtered_df = df[
        #    df["labels"].str.contains(pattern, case=False, na=False, regex=True)
        # ]

        # if not filtered_df.empty:  # Append non-empty dataframes to the list
        #    appended_data.append(filtered_df)

        not_filtered_df = df[
            ~df["labels"].str.contains(pattern, case=False, na=False, regex=True)
            & df["labels"].notna()
            & (df["labels"] != "")
        ]

        if not not_filtered_df.empty:
            not_high_priority.append(not_filtered_df)

        print(f"Processed: {i}")
    except Exception as e:
        print(f"Error processing {i}: {e}")


# appended_data = pd.concat(appended_data, ignore_index=True)
not_high_priority = pd.concat(not_high_priority, ignore_index=True)
```
