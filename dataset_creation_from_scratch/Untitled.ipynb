```python
import pandas as pd
import glob

input_files = glob.glob('all_issue_event/*.csv')
```

```python
all_filtered_rows = []

for file in input_files:
    try:
        df = pd.read_csv(file)
        
        filtered_df = df[df['labels'].str.contains('high', case=False, na=False) & 
                        df['labels'].str.contains('priority', case=False, na=False)]
        
        all_filtered_rows.append(filtered_df)
        
    except pd.errors.EmptyDataError:
        print(f"The file {file} is empty.")
    except pd.errors.ParserError:
        print(f"Error parsing the file {file}. It might be corrupted or not a CSV.")
    except Exception as e:
        print(f"An unexpected error occurred with file {file}. Error: {e}")

# Combine all filtered dataframes
aggregated_df = pd.concat(all_filtered_rows, ignore_index=True)
aggregated_df
```

```python
output_path = "aggregated_filtered_labels.csv"
aggregated_df.to_csv(output_path, index=False)
```
