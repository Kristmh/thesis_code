```python
import dask.dataframe as dd
```

```python
data_debt = dd.read_csv("../TD_dataset_events/issues_debt_*.csv", dtype={'Unnamed: 0': 'float64',
                                                    'id': 'object'}
                        )
```

```python
data_debt1 = data_debt.compute()
```

```python
data_debt1.drop(columns=["Unnamed: 0"] , inplace = True)
```

```python
data_debt1.reset_index(inplace=True)
```

```python
data_debt1
```

```python
data_debt1 = data_debt1.drop_duplicates(subset=['body'], keep='last')
```

```python
data_debt1
```

```python
data_debt1.drop(columns=["index"]).to_csv("all_possible_debt.csv")
```
