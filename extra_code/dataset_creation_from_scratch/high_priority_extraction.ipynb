## Extracting High Priority Issues
This notebook contains the code to filter rows with labels that indicate high priority from the dataset and standardize these labels to "high_priority".


```python
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('all_issue_event/issues_all_2021-03-02.csv')

df.head()
```

```python
# Extract unique labels containing the word "high"
high_labels = df['labels'].dropna().unique()
high_labels = [label for label in high_labels if 'high' in label.lower()]
high_labels
```

```python
# Calculate the number of occurrences for each label containing the word "high"
label_counts = df[df['labels'].isin(high_labels)]['labels'].value_counts()

# Sort the labels by their count
sorted_label_counts = label_counts.sort_values(ascending=False)
sorted_label_counts
```
