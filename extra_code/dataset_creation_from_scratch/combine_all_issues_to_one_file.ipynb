# Combine all issues into one CVS file

```python
import pandas as pd
import os
```

```python
# Step 1: List all the CSV files you want to combine
# Replace 'path_to_your_directory' with the directory containing your CSV files
path = 'all_issue_event/'
all_files = [os.path.join(path, f) for f in os.listdir(path) if f.endswith('.csv')]
```

```python
dfs = []
# Read each CSV file into a DataFrame one by one
for file in all_files:
    try:
        print(f"Reading file: {file}")
        df = pd.read_csv(file, dtype=str)
        dfs.append(df)
    except Exception as e:
        print(f"Error reading file {file}. Error: {e}")
```

```python
# Concatenate all the DataFrames
combined_df = pd.concat(dfs, ignore_index=True)
combined_df
```

```python
# Save the concatenated DataFrame to a new CSV file
combined_df.to_csv('all_issue_events.csv', index=False)
```
